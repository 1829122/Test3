```python
def factadressCall(stn):
  url = 'http://apis.data.go.kr/B552584/cleansys/recentThYearCmprFyerBsnesStatsInfo'
  params ={'serviceKey' : '서비스키', 'areaNm' : stn, 'searchYear' : '2018', 'type' : 'json' }

  response = requests.get(url, params=params)
  data2 = response.json()
  return data2
fact_data1 = factadressCall('울산') #울산
fact_data1
```

```python
fact_data2 = factadressCall('경상북도') #울산
fact_data2
```

```python
# 가정된 추출된 데이터
Ulsan_data = [
    {
        "fact_manage_nm": item["fact_manage_nm"],
        "address" : item["fact_adres"]
    } for item in fact_data1["response"]["body"]["items"]
]

Pohang_data = [
    {
        "fact_manage_nm": item["fact_manage_nm"],
        "address" : item["fact_adres"]
    } for item in fact_data2["response"]["body"]["items"] if '포항' in item["fact_adres"]
]


# "울산광역시" 키 아래에 데이터 넣기
reformatted_data = {
    "울산광역시": Ulsan_data,
    "포항시" : Pohang_data
}

# JSON 파일로 저장

with open('data2.json', "w", encoding='utf-8') as json_file:
    json.dump(reformatted_data, json_file, ensure_ascii=False)

reformatted_data["울산광역시"][0]["address"]
```
