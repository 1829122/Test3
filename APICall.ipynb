```python
import requests
import json

def factadressCall(stn):
  url = 'http://apis.data.go.kr/B552584/cleansys/recentThYearCmprFyerBsnesStatsInfo'
  params ={'serviceKey' : '서비스키', 'areaNm' : stn, 'searchYear' : '2018', 'type' : 'json' }
  try:
    response = requests.get(url, params=params)
    data2 = response.json()
    return data2
  except json.JSONDecodeError:
        factadressCall(stn)

fact_data1 = factadressCall('울산') #울산
fact_data1
```

```python
import pandas as pd

# 초기 데이터 생성
data = {
    'Time': ['10:00', '13:00'],
    'Temperature': [20, 23]
}

# DataFrame 생성 및 설정
df = pd.DataFrame(data)
df['Time'] = pd.to_datetime(df['Time'])
df.set_index('Time', inplace=True)

# 30분 간격으로 리샘플링 후 선형 보간
df_interpolated = df.resample('30T').interpolate()

# 10분 간격으로 리샘플링
df_resampled = df_interpolated.resample('10T').ffill()

# 결과 출력
print(df_resampled)
```
